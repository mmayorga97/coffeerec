### Observations

For this simple content-based recommender we'll leverage coffee's attributes

- Remove Columns:
  - ID: Noise data, not useful.
  - Unnamed: Noise data, not useful.
  - ICO Number: Seems like a personal ID, irregularly used among the producers. This is the column with more missing values, not useful.
  - Number of Bags: Irrelevant for coffee attributes
  - Bag Weight: Irrelevant attributes
  - uniformity: The value is the same across all rows (value = 10).
  - clean cup: The value is the same across all rows (value = 10).
  - sweetness: The value is the same across all rows (value = 10).
  - defects: The value is the same across all rows (value = 0).
  - Harvest Year: its the same year period for all rows(value = 2022)
  - Status: The value is the same across all rows (value = 'completed').
  - Certification info ('Expiration','Certification Body', 'Certification Address', 'Certification Contact','Category Two Defects','Category One Defects'), not related to the coffee's attributes.

- Data Treatment in columns:
  - adjust column names: lower and underscore for proper name handling.
  - altitude: The column is a string used as a range eg: "a - b", can be split and use a mean of both values if used.
  - color : There are several variations of "yellow green" that needs to be fixed for proper encoding.
  - Processing method: standarize some values from'semi-lavado' to "semi washed".
  - fill empty values with 'unknown'.

# Preprocess

```python
# imports
import pandas as pd
import numpy as np

# load file
coffee_data = pd.read_csv('app/coffee_may2023.csv')

# remove columns
coffee_prep_data = coffee_data.drop(['Unnamed: 0', 'ID','ICO Number','Uniformity','Number of Bags','Harvest Year','Grading Date', 'Status', 'Bag Weight','Clean Cup', 'Sweetness','Category One Defects', 'Defects','Expiration','Certification Body', 'Certification Address', 'Certification Contact','Category Two Defects'],axis=1)

# adjust column names to lower replacing empty spaces with underscore.
coffee_prep_data.columns = coffee_prep_data.columns.str.lower()
coffee_prep_data.columns = coffee_prep_data.columns.str.replace(' ','_')
coffee_prep_data.columns = coffee_prep_data.columns.str.replace('-','_')

# lower values
for col in coffee_prep_data.select_dtypes(include=['object']).columns:
    coffee_prep_data[col] = coffee_prep_data[col].str.lower()

# adjusting data values for proper categorization
coffee_prep_data.processing_method = coffee_prep_data.processing_method.str.replace('semi-lavado','semi washed')
coffee_prep_data.color = coffee_prep_data.color.str.replace('yellow green','yellow-green')
coffee_prep_data.color = coffee_prep_data.color.str.replace('yellow- green','yellow-green')
coffee_prep_data.color = coffee_prep_data.color.str.replace('yello-green','yellow-green')
coffee_prep_data.color = coffee_prep_data.color.str.replace(' ','-')

# filling emptys with "unknown"
coffee_prep_data.variety.fillna('unknown', inplace=True)
coffee_prep_data.processing_method.fillna('unknown', inplace=True)
coffee_prep_data.mill.fillna('unknown', inplace=True)

# function to calculate mean to be used for altitude values, from string to int
def calculate_mean(value):
    value = value.replace('~', '-').replace('a', '-').replace('n','0').replace(' ','') #clean up string
    if '-' in value:
        min_val, max_val = value.split('-')
        return np.mean([int(min_val), int(max_val)])
    else:
        return int(value)
coffee_prep_data.altitude = coffee_prep_data.altitude.astype(str).apply(calculate_mean)

# preparing the data to be used
coffee_prep_data= coffee_prep_data[['lot_number','producer','mill','country_of_origin','variety','processing_method', 'aroma', 'flavor', 'aftertaste', 'acidity', 'body',
'balance', 'overall', 'total_cup_points', 'moisture_percentage','quakers', 'color','altitude']]

# save file
coffee_prep_data.to_csv('app/coffee_may2023_prep.csv', index=False)
```
